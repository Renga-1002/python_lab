# <h3>Question-1 Retrieve data from web page using URLLIB and print the frequency of words from that page</h3>

```python
import urllib.request
from bs4 import BeautifulSoup
from collections import Counter
import re

# Function to fetch the web page content
def fetch_webpage(url):
    try:
        response = urllib.request.urlopen(url)
        html = response.read()
        return html
    except Exception as e:
        print("Error fetching the web page:", str(e))
        return None

# Function to parse HTML and extract text
def extract_text_from_html(html):
    soup = BeautifulSoup(html, 'html.parser')
    text = soup.get_text()
    return text

# Function to calculate word frequencies
def calculate_word_frequencies(text):
    # Tokenize the text into words
    words = re.findall(r'\b\w+\b', text.lower())
    
    # Calculate word frequencies
    word_freq = Counter(words)
    
    return word_freq

# Main function
if __name__ == "__main__":
    url = 'https://www.geeksforgeeks.org/' # Replace with the URL of the web page you want to analyze
    
    # Fetch the web page content
    html_content = fetch_webpage(url)
    
    if html_content:
        # Extract text from HTML
        page_text = extract_text_from_html(html_content)
        
        # Calculate word frequencies
        word_frequencies = calculate_word_frequencies(page_text)
        
        # Print word frequencies
        for word, frequency in word_frequencies.items():
            print(f"{word}: {frequency}")
    else:
        print("Web page content not retrieved.")
```

# <h3>question 2: Retreive and display all hyperlinks (ie., HREF attribute) from a webpage using Beautiful soup</h3>

```python
import urllib.request
from bs4 import BeautifulSoup

# Function to fetch the web page content
def fetch_webpage(url):
    try:
        response = urllib.request.urlopen(url)
        html = response.read()
        return html
    except Exception as e:
        print("Error fetching the web page:", str(e))
        return None

# Function to extract and display hyperlinks
def display_hyperlinks(html):
    soup = BeautifulSoup(html, 'html.parser')
    
    # Find all <a> tags (anchor tags)
    links = soup.find_all('a')
    
    # Iterate through the <a> tags and display their href attributes
    for link in links:
        href = link.get('href')
        if href:
            print(href)

# Main function
if __name__ == "__main__":
    url = 'https://www.geeksforgeeks.org/'  # Replace with the URL of the web page you want to analyze
    
    # Fetch the web page content
    html_content = fetch_webpage(url)
    
    if html_content:
        # Display hyperlinks
        display_hyperlinks(html_content)
    else:
        print("Web page content not retrieved.")
```

# <h3>Question3: Create a HTML file for the following student marks and print the number of students and their names and marks</h3>

```python
# Define student data as a list of dictionaries
student_data = [
    {'ID': 'DS01', 'Name': 'Rex', 'mark1': 87, 'mark2': 57, 'mark3': 74},
    {'ID': 'DS02', 'Name': 'Peter', 'mark1': 68, 'mark2': 98, 'mark3': 55}
]

# Function to create an HTML table from student data
def create_html_table(data):
    table_html = "<table border='1'>\n"
    table_html += "<tr><th>ID</th><th>Name</th><th>Mark 1</th><th>Mark 2</th><th>Mark 3</th></tr>\n"
    
    for student in data:
        table_html += f"<tr><td>{student['ID']}</td><td>{student['Name']}</td><td>{student['mark1']}</td><td>{student['mark2']}</td><td>{student['mark3']}</td></tr>\n"
    
    table_html += "</table>"
    return table_html

# Write HTML content to a file
html_content = create_html_table(student_data)

with open("student_marks.html", "w") as html_file:
    html_file.write(html_content)

print("HTML file 'student_marks.html' has been created.")

# Read and parse the HTML file to print the number of students and their details
from bs4 import BeautifulSoup

with open("student_marks.html", "r") as html_file:
    soup = BeautifulSoup(html_file, "html.parser")
    
    # Find all table rows (excluding the header row)
    rows = soup.find_all("tr")[1:]
    
    student_count = 0
    for row in rows:
        columns = row.find_all("td")
        if len(columns) == 5:
            student_count += 1
            student_id = columns[0].get_text()
            student_name = columns[1].get_text()
            student_mark1 = columns[2].get_text()
            student_mark2 = columns[3].get_text()
            student_mark3 = columns[4].get_text()
            print(f"ID: {student_id}, Name: {student_name}, Mark 1: {student_mark1}, Mark 2: {student_mark2}, Mark 3: {student_mark3}")

print(f"Number of students: {student_count}")
```

# <h3>Question4. Create a JSON file for the following Students Marks and print the number of students and their names and marks</h3>

```python
import json

# Define student data as a list of dictionaries
student_data = [
    {'ID': 'DS01', 'Name': 'Rex', 'Semester1': [80, 55], 'Semester2': [50, 70, 82]},
    {'ID': 'DS02', 'Name': 'Peter', 'Semester1': [92, 75]}
]

# Step 1: Write data to a JSON file
with open("student_marks.json", "w") as json_file:
    json.dump(student_data, json_file)

print("JSON file 'student_marks.json' has been created.")

# Step 2: Read data from the JSON file
with open("student_marks.json", "r") as json_file:
    student_data = json.load(json_file)

    # Step 3: Print the number of students and their details
    student_count = 0
    for student in student_data:
        student_count += 1
        student_id = student['ID']
        student_name = student['Name']
        semester_1marks = student.get('Semester1',[])
        semester_2marks = student.get('Semester2',[])
        print(f"ID: {student_id}, Name: {student_name}")
        print(f"Semester 1 Marks: {', '.join(map(str, semester_1marks))}")
        print(f"Semester 2 Marks: {', '.join(map(str, semester_2marks))}")
        print()

print(f"Number of students: {student_count}")
```
